```python
# Importar bibliotecas necesarias
import pandas as pd
import altair as alt

# Cargar el archivo CSV con el delimitador correcto (en este caso, ';')
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/rmcadenas.csv', encoding='latin-1', sep=';')

# Imprimir las columnas del DataFrame para verificar los nombres
print(data.columns)

# Limpiar los nombres de las columnas, eliminando posibles espacios extra
data.columns = data.columns.str.strip()

# Verificar nuevamente las columnas después de limpiar los nombres
print(data.columns)

# Asegúrate de que los nombres de las columnas en 'columnas_relevantes' coincidan exactamente con los nombres en el DataFrame
# Ajusta los nombres en 'columnas_relevantes' si es necesario
columnas_relevantes = ['Cadena', 'Región Metropolitana']

# Seleccionar las columnas relevantes
datos_filtrados = data[columnas_relevantes].dropna()

# Crear un gráfico de barras para visualizar las Ventas Anuales por Cadena
# agregué color='#e41a1c' para cambiar el color de las barras, puedes elegir algún color de su paleta!

chart = alt.Chart(datos_filtrados).mark_bar(color='#e41a1c').encode(
    x=alt.X('Cadena:N', sort='-y', title='Cadena'),
    y=alt.Y('Región Metropolitana:Q', title='Región Metropolitana'),
    tooltip=['Cadena', 'Región Metropolitana']
).properties(
    title='Locales en la Región Metropolitana por Cadena',
    width=600,
    height=400
)

chart
```

```python
# Crear un gráfico de barras para visualizar las Ventas Anuales por Cadena
chart = alt.Chart(datos_filtrados).mark_bar().encode(
    x=alt.X('Cadena:N', sort='-y', title='Cadena'),
    y=alt.Y('Región Metropolitana:Q', title='Región Metropolitana'),
    tooltip=['Cadena', 'Región Metropolitana']
).properties(
    title='Locales en la Región Metropolitana por Cadena',
    width=600,
    height=400
)

chart
# Guardar el gráfico como archivo HTML para visualizarlo en un navegador
chart.save('Visualización03.html')

# Mostrar un mensaje indicando que el archivo ha sido creado
print("El gráfico se ha guardado como 'Visualización03.html'. Ábrelo en un navegador para visualizarlo.")
```
