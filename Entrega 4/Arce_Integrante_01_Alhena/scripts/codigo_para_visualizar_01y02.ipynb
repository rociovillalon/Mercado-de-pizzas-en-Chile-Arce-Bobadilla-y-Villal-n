```python
# Importar bibliotecas necesarias
import pandas as pd
import altair as alt

# Cargar el archivo CSV con el delimitador correcto (en este caso, ';')
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/tablalimpiaaa.csv', encoding='latin-1', sep=';')

# Imprimir las columnas del DataFrame para verificar los nombres
print(data.columns)

# Limpiar los nombres de las columnas, eliminando posibles espacios extra
data.columns = data.columns.str.strip()

# Verificar nuevamente las columnas después de limpiar los nombres
print(data.columns)

# Asegúrate de que los nombres de las columnas en 'columnas_relevantes' coincidan exactamente con los nombres en el DataFrame
# Ajusta los nombres en 'columnas_relevantes' si es necesario
columnas_relevantes = ['Año', 'Total n° de empresas']

# Seleccionar las columnas relevantes
datos_filtrados = data[columnas_relevantes].dropna()

# Crear un gráfico de barras para visualizar las Ventas Anuales por Cadena
chart = alt.Chart(datos_filtrados).mark_bar(color='#ffdf00').encode(
    x=alt.X('Año:O', sort='ascending', title='Año'),
    y=alt.Y('Total n° de empresas:Q', title='Total n° de empresas en el rubro de Actividades de restaurantes y de servicio móvil de comidas'),
    tooltip=['Año', 'Total n° de empresas']
).properties(
    title='Total n° de empresas por Año',
    width=600,
    height=400
    )

chart
```

```python
# Importar bibliotecas necesarias
import pandas as pd
import altair as alt

# Cargar el archivo CSV con el delimitador correcto (en este caso, ';')
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/tablalimpiaaa.csv', encoding='latin-1', sep=';')

# Imprimir las columnas del DataFrame para verificar los nombres
print(data.columns)

# Limpiar los nombres de las columnas, eliminando posibles espacios extra
data.columns = data.columns.str.strip()

# Verificar nuevamente las columnas después de limpiar los nombres
print(data.columns)

# Asegúrate de que los nombres de las columnas en 'columnas_relevantes' coincidan exactamente con los nombres en el DataFrame
# Ajusta los nombres en 'columnas_relevantes' si es necesario
columnas_relevantes = ['Año', 'Total Ventas anuales en UF']

# Seleccionar las columnas relevantes
datos_filtrados = data[columnas_relevantes].dropna()
# La columna 'Total Ventas anuales en UF' tiene comas (,) y no e sun valor númérico para Pandas, tenemos que converturla a númerico reemplazando esas comas (,) por puntos (.) para que lo entienda como decimal de forma correcta

datos_filtrados['Total Ventas anuales en UF'] = datos_filtrados['Total Ventas anuales en UF'].str.replace(',', '.').astype(float)

datos_filtrados.head(5)

# Crear un gráfico de barras para visualizar las Ventas Anuales por Cadena
chart = alt.Chart(datos_filtrados).mark_bar(color='#ff7f00').encode(
    x=alt.X('Año:O', sort='ascending', title='Año'),
    y=alt.Y('Total Ventas anuales en UF:Q', title='Total Ventas anuales en UF en el rubro de Actividades de restaurantes y de servicio móvil de comidas'),
    tooltip=['Año', 'Total Ventas anuales en UF']
).properties(
    title='Total Ventas anuales en UF por Año',
    width=600,
    height=400
    )

chart
```
