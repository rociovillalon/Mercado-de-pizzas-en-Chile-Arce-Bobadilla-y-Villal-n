```python
import pandas as pd

# Cargar y revisar el archivo
file_path = '/content/drive/MyDrive/grafica/localesreviewsyratingfinal.csv'

#Cambio en el formato del archivo para su visualización
data = pd.read_csv(file_path, encoding='latin-1')

# Dividir los datos en columnas usando el delimitador ';'
data = pd.read_csv(file_path, encoding='latin1', delimiter=';')

# Mostrar las primeras filas para verificar la estructura
data.head()

import matplotlib.pyplot as plt

# Datos de la base de datos
Comunas= ['Santiago', 'Maipú', 'Puente Alto', 'Las Condes', 'La Florida', 'Ñuñoa', 'Providencia', 'San Miguel', 'La Cisterna', 'Peñalolén', 'Huechuraba', 'La Reina', 'Lo Barnechea', 'Macul', 'Lo Prado', 'Quilicura', 'Independencia', 'Cerrillos']
Rating= [5.0, 3.4, 3.7, 1.0, 3.9, 2.7, 0.0, 2.8, 4.4, 4.1, 5.0, 4.8, 3.7, 4.9, 4.2, 4.1, 4.1, 4.2]

# Crear gráfico de barras horizontales
plt.figure(figsize=(10, 5))
plt.barh(Comunas, Rating, color='green', edgecolor='black')
plt.title('Rating por cadena Pizzapizza')
plt.xlabel('Rating')
plt.ylabel('Comunas')
plt.show()
```
