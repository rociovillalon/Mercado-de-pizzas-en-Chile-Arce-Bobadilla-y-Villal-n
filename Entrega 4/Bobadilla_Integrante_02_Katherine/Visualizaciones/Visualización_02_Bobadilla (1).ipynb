```python
import pandas as pd

# Cargar y revisar el archivo
file_path = '/content/drive/MyDrive/grafica/localesreviewsyratingfinal.csv'
# file_path = '/content/drive/MyDrive/grafica/localesreviewsyratingfinal.csv'

#Cambio en el formato del archivo para su visualización
data = pd.read_csv(file_path, encoding='latin-1')

# Dividir los datos en columnas usando el delimitador ';'
data = pd.read_csv(file_path, encoding='latin1', delimiter=';')

# Mostrar las primeras filas para verificar la estructura
data.head()
```

```python
# Tranformar los valores Nan en Rating en 0 para que no se descarten después Providencia es el caso para Pizzapizza.
# Asumiremos que si es 0, es porque no hay datos.

data['Rating'] = data['Rating'].fillna(0)
```

```python
# Seleccionamos solo la Cadena 'Pizzapizza' en los datos y también solo las columnas definidas.


data_filtrada = data[data['Cadena'] == 'Pizzapizza']

data_filtrada
```

```python
import altair as alt

chart = alt.Chart(data_filtrada).mark_bar(color='green').encode(
    x='Rating:Q',
    y="Comuna:N"
).properties(
    width=600,
    title='Rating por cadena Pizzapizza'
    )

chart
```

```python
# exportar a html
chart.save('pizzapizza_rating.html')
```
